### Mengatasi/menghilangkan data yang kosong

### Data kategorik yang kosong akan diisi dengan modus

```python
for data_kategorik in ['Lead Source','Last Activity','Country','Specialization','How did you hear about Madugital','What is your current occupation','What matters most to you in choosing a product','Tags','Lead Quality','Lead Profile','City','Asymmetrique Activity Index','Asymmetrique Profile Index']:
    data[data_kategorik] = data[data_kategorik].fillna(data[data_kategorik].value_counts().idxmax())


```

```python
data.info()
data.isna().mean()*100
```

### Data numerik yang kosong akan diisi dengan rata-rata(mean)

```python
for data_numerik in ['TotalVisits','Page Views Per Visit','Asymmetrique Activity Score','Asymmetrique Profile Score']:
    data[data_numerik] = data[data_numerik].fillna(data[data_numerik].mean())
data.info()
data.isna().mean()*100
```

```python
new_df_cap_4 = new_df_cap_3.copy()

new_df_cap_4['Asymmetrique Profile Score'] = np.where(
    new_df_cap_4['Asymmetrique Profile Score'] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap_4['Asymmetrique Profile Score'] < lower_limit,
        lower_limit,
        new_df_cap_4['Asymmetrique Profile Score']
    )
)

new_df_cap_4.shape
```

```python
from sklearn.model_selection import cross_val_score

# List of max_depth values to try
max_depths = range(1, 50)

# List of min_samples_split values to try
min_samples_splits = range(2, 50)

# Initialize lists to store mean accuracy scores and standard deviations
mean_scores = []
std_scores = []

# Loop through max_depth and min_samples_split combinations and perform cross-validation
for max_depth in max_depths:
    for min_samples_split in min_samples_splits:
        # Create decision tree classifier
        clf = DecisionTreeClassifier(max_depth=max_depth, min_samples_split=min_samples_split)

        # Perform cross-validation
        scores = cross_val_score(clf, X_train, y_train, cv=5)

        # Store mean and standard deviation of scores
        mean_scores.append(np.mean(scores))
        std_scores.append(np.std(scores))

# Find the best combination of max_depth and min_samples_split based on mean accuracy score
best_idx = np.argmax(mean_scores)
best_max_depth = max_depths[best_idx // len(min_samples_splits)]
best_min_samples_split = min_samples_splits[best_idx % len(min_samples_splits)]
best_mean_score = np.max(mean_scores)
best_std_score = std_scores[best_idx]

# Print the results
print(f"Best combination of max_depth and min_samples_split:")
print(f"Max Depth: {best_max_depth}")
print(f"Min Samples Split: {best_min_samples_split}")
print(f"Mean accuracy score: {best_mean_score}")
print(f"Standard deviation: {best_std_score}")
```
